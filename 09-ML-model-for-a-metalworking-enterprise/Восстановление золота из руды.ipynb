## Введение

Цель: построить прототип модели машинного обучения для компании "Цифра". Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды, что в дальнейшем поможет оптимизировать производство и исключит запуск предприятий с убыточными характеристиками.

Дано: параметры добычи и очистки золотосодержащей руды, индексированные датой и временем получения информации.

Наименование данных:

* Rougher feed — исходное сырье
* Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
* Xanthate — ксантогенат (промотер, или активатор флотации);
* Sulphate — сульфат (на данном производстве сульфид натрия);
* Depressant — депрессант (силикат натрия).
* Rougher process (англ. «грубый процесс») — флотация
* Rougher tails — отвальные хвосты
* Float banks — флотационная установка
* Cleaner process — очистка
* Rougher Au — черновой концентрат золота
* Final Au — финальный концентрат золота

Параметры этапов:

* air amount — объём воздуха
* fluid levels — уровень жидкости
* feed size — размер гранул сырья
* feed rate — скорость подачи

Наименование этапов:

* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики

Наименование параметров сырья в каждом блоке:

* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики

Целевые признаки:

* эффективность обогащения чернового концентрата rougher.output.recovery;
* эффективность обогащения финального концентрата final.output.recovery.

Решение о признаках, участвующих в обучении модели необходимо принять самостоятельно в ходе работы.

Порядок работы:

1 - подготовить данные, оценить корректность расчета эффективности обогащения, проанализировать признаки, отсутствующие в выборке для теста;

2 - провести исследовательский анализ данных: 
- изменения концентрации металлов (Au, Ag, Pb) на различных этапах очистки,
- распределения размеров гранул сырья на обучающей и тестовой выборках;
- суммарной концентрации всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах.

3 - построить и обучить модель, оценить качество модели при помощи кроссвалидации, использовать метрику SMAP, протестировать модель.

### Проверка эффективности обогащения

Выполним проверку эффективности обогащения, вычисляемой по формуле:
    
    *RECOVERY = C*(F-T)/F/(C-T)*100%*

где:
C — доля золота в концентрате после флотации/очистки;
F — доля золота в сырье/концентрате до флотации/очистки;
T — доля золота в отвальных хвостах после флотации/очистки.

```python
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']
```

Запишем проверку по формуле в новый столбец *check*

```python
data_train['check'] = C*(F-T)/F/(C-T)*100
data_train['check']
```

Вычислим MAE с помощью модуля sklearn

```python
mae = mean_absolute_error(data_train['rougher.output.recovery'],data_train['check']).round(1)
mae
```

Вывод:

MAE - средняя абсолютная ошибка, которая показывает как сильно предсказания в среднем отличаются от истинных значений. Полученное значение округлено до одной цифры после запятой и равно 0, а значит расчеты верны.

### Исследование суммарной концентрации всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах

```python
sns.reset_orig()
```

```python
columns = ['rougher.input.feed_au','rougher.input.feed_ag','rougher.input.feed_pb','rougher.input.feed_sol',\
          'rougher.output.concentrate_au','rougher.output.concentrate_ag','rougher.output.concentrate_pb',\
          'rougher.output.concentrate_sol','final.output.concentrate_au','final.output.concentrate_ag',\
          'final.output.concentrate_pb','final.output.concentrate_sol']
```

```python
table = df_0.pivot_table(index='date_new', values = columns, aggfunc = 'median').reset_index()
```

```python
state = ['rougher.input.feed','rougher.output.concentrate','final.output.concentrate']
st_dict = {'rougher.input.feed':'сырье','rougher.output.concentrate':'черн.конц.','final.output.concentrate':'фин.'}
```

```python
for x in state:
    z = table['date_new']
    y1 = table[f'{x}_au']
    y2 = table[f'{x}_ag']
    y3 = table[f'{x}_pb']
    y4 = table[f'{x}_sol']

    fig, ax = plt.subplots()

    p1 = ax.bar(z, y1,color='gold')
    p2 = ax.bar(z, y2, bottom = y1,color='silver')
    p3 = ax.bar(z, y3, bottom = y1+y2,color='lightblue')
    p4 = ax.bar(z, y4, bottom = y1+y2+y3,color='navy')

    fig.set_figwidth(20)
    fig.set_figheight(10)

    plt.title(f'Исследование суммарной концентрации веществ в {st_dict.get(x)}',size=20)
    plt.xlabel('Время исследования', fontsize=15);
    plt.ylabel('Концентрация', fontsize=15);
    plt.tick_params(axis='both', which='major', labelsize=10)
    plt.legend((p1[0], p2[0],p3[0],p4[0]),('Au', 'Ag','Pb','Sol'))
    plt.show()
```

Вывод: мы визуализировали изменение суммарной концентрации веществ Au,Ag,Pb,Sol на этапах сырья, чернового и финального концентратов с помощью столчатой диаграммы, это дало нам возможность оценить изменение каждого металла отдельно.
В первом случае мы видим низкую суммарную концентрацию металлов, более 60% поступающего на переработку сырья занимает твердые примеси Sol. Во втором случае, после флотации суммарная концентрация увеличилась, но дооля твердых примесей еще около 50%. В финале, твердых примесей значительно меньше, около 15%, в то время как доля Au станосится преобладающей - более 60%.

Напишем формулу для вычисления SMAPE симметричного абсолютного процентного отклонения и итогового SMAPE

```python
def smap(target,prediction):
    score = np.sum(np.abs(prediction-target)*2/(np.abs(prediction)+np.abs(target)))*100
    return score/len(target)
```

```python
def final_smap(rougher,final):
    score_final = 0.25*rougher + 0.75*final
    return score_final
```

```python
best_model = None
best_result = 100
best_depth = 0
for depth in range(1, 10):
    model_dtc_r = DecisionTreeRegressor (random_state=state, max_depth=depth)
    model_dtc_r.fit(features_train,target_rougher_train)
    scores_dtc_rougher = cross_val_score (model_dtc_r,features_train,target_rougher_train,cv=5,\
                                         scoring = make_scorer(smap, greater_is_better=False))
    final_score_dtc_rougher = scores_dtc_rougher.mean()
    model_dtc_f = DecisionTreeRegressor (random_state=state, max_depth=depth) 
    model_dtc_f.fit(features_train,target_final_train)
    scores_dtc_final = cross_val_score (model_dtc_f,features_train,target_final_train,cv=5,\
                                       scoring = make_scorer(smap, greater_is_better=False)) 
    final_score_dtc_final = scores_dtc_final.mean()
    result_dtc_smap = final_smap(final_score_dtc_rougher,final_score_dtc_final)*(-1)
    if result_dtc_smap < best_result:
        best_model = model_dtc_r
        best_result = result_dtc_smap
        best_depth = depth
print("Средняя оценка SMAP наилучшей модели DTC для прогноза доли Au в черновом концентрате:", best_result, "Глубина дерева:", best_depth)
```
